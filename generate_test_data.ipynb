```python
import numpy as np

N = 1_000_000

CHUNK = int(N/100)##_000_000
OUT = "world_points.json"

with open(OUT, "w") as f:
    f.write("[\n")

    written = 0

    first = True

    while written < N:
        size = min(CHUNK, N - written)

        lon = np.random.uniform(-180, 180, size)
        lat = np.random.uniform(-90, 90, size)
        val = np.random.randint(1, 3, size)

        for i in range(size):
            if not first:
                f.write(",\n")
            first = False
            f.write(f"[{lon[i]:.6f},{lat[i]:.6f},{val[i]}]")

        written += size

    f.write("\n]")
```
